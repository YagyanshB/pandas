```python
# combine filters together

df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]

df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head()
```

```python
df2.shape
```

```python
# Remove USA from event name

df2.loc[:, 'Event name'] = df2['Event name'].str.split('(').str[0].str.strip()
```

```python
df2.head()
```

```python
df2.shape
```

```python
# Clean up Athlete Age

df2['Athlete age'] = df2['Year of event'] - df2['Athlete year of birth']
```

```python
# remove h from athlete performance

df2['Athlete performance'] = df2['Athlete performance'].str.replace('h', '')
```

```python
df2.head()
```

```python
# Drop Columns

df2 = df2.drop(columns = ['Athlete club', 'Athlete age category', 'Athlete country', 'Athlete year of birth'], axis = 1)
```

```python
df2.head()
```

```python
# Clean Up Null Values

df2.isnull().sum()
```

```python
df2[df2['Athlete age'].isnull() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
# Check for duplicates

df2[df2.duplicated()== True]
```

```python
# Reset Index

df2 = df2.reset_index(drop = True)
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
# Rename Columns

df2 = df2.rename(columns = {'Athlete performance': 'athlete_performance', 'Athlete average speed': 'athlete_average_speed', 'Year of event': 'year',
                            'Event dates': 'race_day', 'Event name': 'race_name', 'Event distance/length': 'race_length', 'Event number of finishers': 'race_number_of_finishers',
                            'Athlete gender': 'athlete_gender', 'Athlete ID': 'athlete_id'})
```

# Investigative Data Analysis & Insights:

```python
# Difference in speed for the 50km, 50mi male to female

df2.groupby(['athlete_gender', 'race_length'])['athlete_average_speed'].mean()
```

```python
# Age groups best in the 50 mile Race (20 + races min)

df2.groupby(['athlete_gender', 'race_length'])['Athlete age'].mean()
```

```python
# Age Groups best in 50 mile race


(df2.query('race_length == "50mi"')
    .groupby('Athlete age')['athlete_average_speed']
    .agg([('mean', 'mean'), ('count', 'count')])
    .sort_values('mean', ascending=False)
    .query('`count` > 19')
).head(10)
```

```python
# Age groups worst in 50 mile race


(df2.query('race_length == "50mi"')
    .groupby('Athlete age')['athlete_average_speed']
    .agg([('mean', 'mean'), ('count', 'count')])
    .sort_values('mean', ascending=True)
    .query('`count` > 19')
).head(10)
```

```python
# Using a lambda function

df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2.head()
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x in [12, 1, 2] else ('Spring' if x in [3, 4, 5] else ('Summer' if x in [6, 7, 8] else 'Fall')))
```

```python
df2.head()
```

```python
(df2.groupby('race_season')['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending=False)
)
```
